## Original Signal

The original, raw sound wave

```python
import matplotlib.pyplot as plt
from math import sin, radians

frames = 2000
signal = [sin(radians(i)) for i in range(frames)]
plt.plot(signal)
plt.show()
```

## Pooling

Basic Pooling with multiple options:
- Min
- Max
- Mean
- Average

```python
import scipy.integrate as integrate

# select pooling type
# min / max / mean / avr
pooling_type = "mean"

# define filter
filter_size = 200
filter = [1 for _ in range(filter_size)]

# apply convolution with filter
step_size = 1
convolved_signal = []
for i in range(0, frames - filter_size, step_size):
    res:float
    match pooling_type:
        case "min":
            res = min(signal[i:i+filter_size])
        case "max":
            res = max(signal[i:i+filter_size])
        case "mean":
            res = sum(signal[i:i+filter_size])/filter_size
        case "avr":
            res = signal[i:i+filter_size][int(filter_size/2)]
    convolved_signal.append(res)

plt.plot(convolved_signal)
plt.ylim([-1, 1])
plt.show()
```

## Convolution

```python
# define filter
filter_size = 100
filter = [0.01*x+0 for x in range(filter_size)]
plt.plot(filter)
plt.show()

# apply convolution with filter
step_size = 1
convolved_signal = []
for i in range(0, frames - filter_size, step_size):
    res = sum([j*k for j, k in zip(signal[i:i+filter_size], filter)])/filter_size
    convolved_signal.append(res)

plt.plot(convolved_signal)
plt.ylim([min(min(convolved_signal), -1), max(max(convolved_signal), 1)])
plt.show()
```
