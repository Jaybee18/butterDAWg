# Distortion

The base signal gets distorted with the given function:


### Algorithm
The signal gets iterated, and for every value in the signal the distortion function gets called with said value and outputs the distorted value.

```python
import matplotlib.pyplot as plt
from math import sin, radians

# generate the signal
frames = 2000
signal = [sin(radians(i)) for i in range(frames)]
plt.plot(signal)
plt.show()
```

```python
# define the distortion function
func = lambda x: x/100
func = lambda x: x*1
func = lambda x: sin(radians(x*0.9))

d = 100
plt.plot([x/d for x in range(-d, d, 1)], [func(x) for x in range(-d, d, 1)])
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.grid(True, which="both")
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")
```

```python
import math

# apply distortion
drive = 5
#test = lambda x: (x - (x**3)/3 if -1 <= x <= 1 else (x/abs(x))*drive) if x != 0 else 0
soft = lambda x: ((math.exp(x*drive)-math.exp(-x*drive))/(math.exp(x*drive)+math.exp(-x*drive))) if x != 0 else 0
step = lambda x: -1 if x < 0 else 1
sine = lambda x: sin(radians(x*500*drive))
distortion_function = soft
distorted_signal = [distortion_function(signal[x]) for x in range(frames)]

plt.plot([x/100 for x in range(-100, 100, 1)], [distortion_function(x/100) for x in range(-100, 100, 1)])
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.grid(True, which="both")
plt.axhline(y=0, color="k")
plt.axvline(x=0, color="k")
plt.title("Distortion function")
plt.show()

plt.plot(distorted_signal)
#plt.xlim([-1, 1])
plt.ylim([-1.1, 1.1])
plt.title("Distorted signal")
plt.show()
```
